```python
import numpy as np
import pandas as pd

# Define the thickness function
def thickness(r):
    return 0.5 / r**2

# Generate r values and corresponding T values
r_values = np.arange(10, 101, 10)  # r values from 10 to 100 in steps of 10
T_values = thickness(r_values)

# Create a Pandas DataFrame
data = pd.DataFrame({
    "Radius (r) [cm]": r_values,
    "Thickness (T) [cm]": T_values
})

# Display the DataFrame
print(data)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the thickness function
def thickness(r):
    return 0.5 / r**2

# Generate r values and corresponding T values
r_values = np.arange(10, 101, 10)  # r values from 10 to 100 in steps of 10
T_values = thickness(r_values)

# Create a Pandas DataFrame
data = pd.DataFrame({
    "Radius (r) [cm]": r_values,
    "Thickness (T) [cm]": T_values
})

# Display the DataFrame (optional)
print(data)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(data["Radius (r) [cm]"], data["Thickness (T) [cm]"], marker='o', color="blue", label="T(r)")
plt.title("Thickness of Balloon Skin vs Radius", fontsize=14)
plt.xlabel("Radius (r) [cm]", fontsize=12)
plt.ylabel("Thickness (T) [cm]", fontsize=12)
plt.grid(alpha=0.5)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend(fontsize=12)
plt.show()
```
